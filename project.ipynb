### Step 2: Load your dataset

```python
import pandas as pd
import altair as alt

data=pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-03-10/tuition_cost.csv",)
```

```python
data=data.rename(columns={'name':'Name','state':'State','state_code':'State_code','type':'Type','degree_length':'Degree_length','room_and_board':'Room_Board','in_state_tuition':'Instate_tuition','in_state_total':'Instate_total','out_of_state_tuition':'Outstate_tuition','out_of_state_total':'Outstate_total'})
data.head()
```
